```python
import pandas as pd

url = "https://raw.githubusercontent.com/artidoro/frank/main/data/human_annotations_sentence.json"
df = pd.read_json(url)
df
```

Taking only the Validation set

```python
df_v = df[df['split']=="valid"]
df_v = df_v.reset_index()
df_v
```

Each sentence is annotated by human evaluators.  Note that there may be disagreement between them.  We can only count a sentence as factually consistent if all 3 evaluators found no errors (NoE).

```python
df_v['summary_sentences_annotations'][0][0]
```

```python
### MODIFY CODE TO MAKE IT WORK
#from summac.model_summac import SummaCZS, SummaCConv

#model_zs = SummaCZS(granularity="sentence", model_name="vitc", device="cuda") # If you have a GPU: switch to: device="cuda"
#model_conv = SummaCConv(models=["vitc"], bins='percentile', granularity="sentence", nli_labels="e", device="cuda", start_file="default", agg="mean")

### ADD CODE FOR SMART
from smart_eval import matching_functions as mf
from smart_eval import summeval_utils as utils
from smart_eval import scorer
from smart_eval.scorer_test import ScorerTest

matcher = mf.chrf_matcher
smart_scorer = scorer.SmartScorer(matching_fn=matcher)

scores = []
```

```python
import time

#evaluations
hash_ = []
model = []
SummaC_zs_eval = []
SummaC_conv_eval = []
SMART_1_eval = []
SMART_2_eval = []
SMART_L_eval = []

start_time = time.time()
for i in range(len(df_v)):
    article = df_v['article'][i] ### ARTICLE
    print ('=====')
    print (i)
    for j in range(len(df_v['summary_sentences'][i])):
        print (j)
        sent = df_v['summary_sentences'][i][j]  ### SENTENCE
        scores = smart_scorer.smart_score(article,sent)
        ### Calculate SummaC
        #score_zs1 = model_zs.score([article], [sent])
        #score_conv1 = model_conv.score([article], [sent])

        ### Calculate SMART
        smart1_results = scores['smart1']['fmeasure']# SMART (article, sent)
        smart2_results = scores['smart2']['fmeasure']# SMART (article, sent)
        smartL_results = scores['smartL']['fmeasure']# SMART (article, sent)
        
        hash_.append( df_v['hash'][i])
        model.append( df_v['model_name'][i])
        #SummaC_zs_eval.append = score_zs1 ### zs score
       # SummaC_conv_eval.append = score_conv1 ### conv score
        
        SMART_1_eval.append(smart1_results) ### SMART 1
        SMART_2_eval.append(smart2_results) ### SMART 2
        SMART_L_eval.append(smartL_results)### SMART L

print ('=====')
print ("--- %s seconds ---" % (time.time()-start_time))      

results = pd.DataFrame()
results['hash'] = hash_
results['model_name'] = model
#results['SummaC_zs'] = SummaC_zs_eval
#results['SummaC_conv'] = SummaC_conv_eval
results['SMART_1_eval'] = SMART_1_eval
results['SMART_2_eval'] = SMART_2_eval
results['SMART_L_eval'] = SMART_L_eval

#file_path = "~/efs/home/zortizl/Hallucination_Research/results/scores-frank-SMART-SummaC.csv"
#results.to_csv('Smart_frank.csv')
```
