```python
import json
with open("dataset.json", "r") as f:
    content = f.read()
dataset = json.loads(content)
```

```python
df = pd.DataFrame(dataset)
```

```python
df
```

```python
matcher = mf.chrf_matcher
smart_scorer = scorer.SmartScorer(matching_fn=matcher)
```

```python
scores = []
for i in range(len(dataset)):
    article = df['wiki_bio_text'][i]
    summary = df['gpt3_text'][i]
    scores.append(smart_scorer.smart_score(article,summary))
```

```python
smart1 = []
smart2 = []
smartL = []

for i in range(len(dataset)):
    smart1.append(scores[i]['smart1']['fmeasure'])
    smart2.append(scores[i]['smart2']['fmeasure'])
    smartL.append(scores[i]['smartL']['fmeasure'])

smart_scores = pd.DataFrame(
    {
        'Smart1': smart1,
        'Smart2': smart2,
        'SmartL': smartL
    })     
```

```python
smart_scores.to_csv('Smart_scores.csv')
```
